```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from prophet import Prophet

# Load the dataset
file_path = 'final_dataset.csv'
df = pd.read_csv(file_path, parse_dates=['Date'])

# Ensure the index is a datetime index
if 'Date' in df.columns:
    df.set_index('Date', inplace=True)

# Focus on these columns for forecasting
target_columns = ['Temperature_C', 'Humidity_%', 'Wind_Speed_kmph', 'UV_Index', 'Atmospheric_Pressure_hPa']

# Function to calculate steps to forecast based on a target date
def calculate_steps_to_forecast(df, target_date):
    last_date = df.index[-1]
    target_date = pd.to_datetime(target_date)

    # Ensure target date is in the future
    if target_date <= last_date:
        raise ValueError(f"Target date {target_date} must be after the last known date {last_date}.")

    return (target_date - last_date).days

# Context-aware noise adjustment
def get_seasonal_noise(column, month):
    if column == 'Rainfall_mm':
        # Low rain noise in winter, higher in monsoon
        if month in [12, 1, 2]:
            return 0.1
        elif month in [7, 8, 9]:
            return 0.5
        else:
            return 0.2
    return 0.3

# Fit Prophet model and forecast future values
def fit_prophet_and_forecast(df, column, steps):
    print(f"Fitting Prophet model for {column}... Forecasting {steps} steps ahead.")

    # Prepare data for Prophet
    series = df[column].interpolate().reset_index()
    series.columns = ['ds', 'y']

    # Fit the Prophet model
    model = Prophet(yearly_seasonality=True)
    model.fit(series)

    # Create future dataframe and forecast
    future = model.make_future_dataframe(periods=steps)
    forecast = model.predict(future)

    # Extract forecasted values for the target range
    forecast_values = forecast[['ds', 'yhat']].tail(steps)

    # Add noise based on season and column
    noise_std = np.std(series['y'].diff().dropna()) * get_seasonal_noise(column, df.index[-1].month)
    forecast_values['yhat'] += np.random.normal(0, noise_std, size=steps)

    # For rainfall in winter, restrict to near zero
    if column == 'Rainfall_mm' and df.index[-1].month in [12, 1, 2]:
        forecast_values['yhat'] = np.clip(forecast_values['yhat'], 0, 1)

    # Plot the results
    plt.figure(figsize=(12, 6))
    plt.plot(df.index[-90:], df[column].iloc[-90:], label='Original Data', color='blue')
    plt.plot(forecast_values['ds'], forecast_values['yhat'], label='Forecast', color='red')
    plt.title(f"Prophet Forecast for {column} with Season-Aware Variability")
    plt.xticks(rotation=45)
    plt.legend()
    plt.tight_layout()
    plt.show()

    return forecast_values['yhat'].values, forecast_values['ds']

# Input: Target date or number of days for prediction
try:
    target_date = input("Enter the target date for prediction (YYYY-MM-DD) or press Enter to skip: ")
    num_days = input("Enter the number of days for forecast or press Enter to skip: ")

    if target_date:
        steps = calculate_steps_to_forecast(df, target_date)
    elif num_days:
        steps = int(num_days)
    else:
        raise ValueError("You must enter either a target date or the number of days for forecast.")

except ValueError as e:
    print(f"Invalid input: {e}")
    exit()

# Forecast for each target column
forecasts = {}
for column in target_columns:
    forecast_values, forecast_dates = fit_prophet_and_forecast(df, column, steps)
    forecasts[column] = forecast_values

# Save forecasted values
forecast_df = pd.DataFrame(forecasts, index=forecast_dates)
forecast_df.to_csv('prophet_forecast.csv')
print("Prophet forecasts saved to 'prophet_forecast.csv'")

# Calculate and display the average values
average_forecast = forecast_df.mean()
print("Average Forecast Values:")
print(average_forecast)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from prophet import Prophet

# Load the dataset
file_path = 'final_dataset.csv'
df = pd.read_csv(file_path, parse_dates=['Date'])

# Ensure the index is a datetime index
if 'Date' in df.columns:
    df.set_index('Date', inplace=True)

# Function to calculate steps to forecast based on a target date
def calculate_steps_to_forecast(df, target_date):
    last_date = df.index[-1]
    target_date = pd.to_datetime(target_date)

    # Ensure target date is in the future
    if target_date <= last_date:
        raise ValueError(f"Target date {target_date} must be after the last known date {last_date}.")

    return (target_date - last_date).days

# Function to add seasonal noise for rainfall
def get_rainfall_noise(month):
    if month in [12, 1, 2]:  # Winter - Minimal rainfall
        return 0.5
    elif month in [7, 8, 9]:  # Monsoon - High rainfall variability
        return 50  # Increased to reflect high monsoon values
    else:  # Other months - Moderate variability
        return 2

# Fit and forecast rainfall using Prophet
def fit_prophet_for_rain(df, target_date):
    steps = calculate_steps_to_forecast(df, target_date)
    print(f"Fitting Prophet model for Rainfall_mm... Forecasting {steps} steps ahead.")

    # Handle zero-inflated data with log transformation
    df['Rainfall_mm_transformed'] = np.log1p(df['Rainfall_mm'])

    # Prepare data for Prophet
    rain_data = df.reset_index()[['Date', 'Rainfall_mm_transformed']]
    rain_data.columns = ['ds', 'y']

    # Add seasonal indicators
    rain_data['monsoon'] = rain_data['ds'].dt.month.isin([7, 8, 9])

    # Initialize and fit the model
    model = Prophet(yearly_seasonality=False, seasonality_mode='multiplicative')

    # Add custom seasonalities
    model.add_seasonality(name='monthly', period=30.5, fourier_order=5)
    model.add_seasonality(name='monsoon', period=365.25, fourier_order=10, condition_name='monsoon')

    # Fit the model
    model.fit(rain_data)

    # Create future dataframe
    future = model.make_future_dataframe(periods=steps)

    # Add monsoon condition
    future['monsoon'] = future['ds'].dt.month.isin([7, 8, 9])

    # Forecast
    forecast = model.predict(future)

    # Reverse log transformation
    forecast['yhat'] = np.expm1(forecast['yhat'])

    # Add seasonal noise dynamically across the forecast
    forecast['month'] = forecast['ds'].dt.month
    forecast['yhat'] += forecast['month'].apply(lambda m: np.random.normal(0, get_rainfall_noise(m)))

    # Ensure values are non-negative
    forecast['yhat'] = np.clip(forecast['yhat'], 0, None)

    # Ensure monsoon values are realistically scaled
    forecast.loc[forecast['month'].isin([7, 8, 9]), 'yhat'] = np.clip(forecast['yhat'], 0, 150)

    # Clip rainfall in winter to near zero
    forecast.loc[forecast['month'].isin([12, 1, 2]), 'yhat'] = np.clip(forecast['yhat'], 0, 2)

    # Plot results (Only last 90 steps of the original data and forecast)
    plt.figure(figsize=(12, 6))
    plt.plot(df.index[-90:], df['Rainfall_mm'].iloc[-90:], label='Original Data', color='blue')
    plt.plot(forecast['ds'][-steps:], forecast['yhat'][-steps:], label='Forecast', color='red')
    plt.title("Prophet Forecast for Rainfall_mm with Season-Aware Variability")
    plt.xlabel("Date")
    plt.ylabel("Rainfall (mm)")
    plt.xticks(rotation=45)
    plt.legend()
    plt.tight_layout()
    plt.show()

    return forecast[['ds', 'yhat']]

# Input: Target date for prediction (YYYY-MM-DD)
target_date = input("Enter the target date for prediction (YYYY-MM-DD): ")

# Forecast Rainfall
rainfall_forecast = fit_prophet_for_rain(df, target_date)

# Save the forecast
rainfall_forecast.to_csv('prophet_rainfall_forecast.csv', index=False)
print("Prophet rainfall forecast saved to 'prophet_rainfall_forecast.csv'")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from prophet import Prophet

# Load the dataset
file_path = 'final_dataset.csv'
df = pd.read_csv(file_path, parse_dates=['Date'])

# Ensure the index is a datetime index
if 'Date' in df.columns:
    df.set_index('Date', inplace=True)

# Function to calculate steps to forecast based on a target date
def calculate_steps_to_forecast(df, target_date):
    last_date = df.index[-1]
    target_date = pd.to_datetime(target_date)

    # Ensure target date is in the future
    if target_date <= last_date:
        raise ValueError(f"Target date {target_date} must be after the last known date {last_date}.")

    return (target_date - last_date).days

# Fit and forecast Rain_Probability_% using Prophet
def fit_prophet_for_rain_probability(df, target_date):
    steps = calculate_steps_to_forecast(df, target_date)
    print(f"Fitting Prophet model for Rain_Probability_%... Forecasting {steps} steps ahead.")

    # Prepare data for Prophet
    rain_prob_data = df.reset_index()[['Date', 'Rain_Probability_%']]
    rain_prob_data.columns = ['ds', 'y']

    # Initialize and fit the model
    model = Prophet(yearly_seasonality=True, seasonality_mode='additive')

    # Add custom seasonalities
    model.add_seasonality(name='monthly', period=30.5, fourier_order=5)

    # Fit the model
    model.fit(rain_prob_data)

    # Create future dataframe
    future = model.make_future_dataframe(periods=steps)

    # Forecast
    forecast = model.predict(future)

    # Ensure probabilities are within 0-100%
    forecast['yhat'] = np.clip(forecast['yhat'], 0, 100)

    # Plot results (Only last 90 steps of the original data and forecast)
    plt.figure(figsize=(12, 6))
    plt.plot(df.index[-120:], df['Rain_Probability_%'].iloc[-120:], label='Original Data', color='green')
    plt.plot(forecast['ds'][-steps:], forecast['yhat'][-steps:], label='Forecast', color='orange')
    plt.title("Prophet Forecast for Rain Probability (%)")
    plt.xlabel("Date")
    plt.ylabel("Rain Probability (%)")
    plt.xticks(rotation=45)
    plt.legend()
    plt.tight_layout()
    plt.show()

    return forecast[['ds', 'yhat']]

# Input: Target date for prediction (YYYY-MM-DD)
target_date = input("Enter the target date for prediction (YYYY-MM-DD): ")

# Forecast Rain Probability
rain_prob_forecast = fit_prophet_for_rain_probability(df, target_date)
```
