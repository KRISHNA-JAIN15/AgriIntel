```python
import pandas as pd
import numpy as np
import pickle

class CropRecommender:
    def __init__(self, average_yields, market_prices, seed_costs, 
                 maintenance_costs, fertilizer_costs, growing_period):
        """
        Initialize the crop recommender with economic data
        """
        self.average_yields = average_yields
        self.market_prices = market_prices
        self.seed_costs = seed_costs
        self.maintenance_costs = maintenance_costs
        self.fertilizer_costs = fertilizer_costs
        self.growing_period = growing_period
        
        # Store the crop means data for soil and climate requirements
        self.crop_means = None
    
    def set_crop_means(self, crop_means):
        """
        Set the optimal growing conditions for each crop
        """
        self.crop_means = crop_means
    
    def calculate_fertilizer_cost(self, crop, soil_n, soil_p, soil_k):
        """
        Calculate fertilizer costs based on soil deficiency compared to crop needs
        
        Parameters:
        - crop: The crop name
        - soil_n, soil_p, soil_k: Actual soil nutrient levels
        
        Returns:
        - Fertilizer cost in ₹ per hectare
        """
        if self.crop_means is None:
            return 0
            
        # Get optimal nutrient levels for the crop using column indices
        # The first three columns should be N, P, K in that order
        try:
            # Assuming the first 3 columns are N, P, K in that order
            optimal_n = self.crop_means.loc[crop].iloc[0]  # First column (N)
            optimal_p = self.crop_means.loc[crop].iloc[1]  # Second column (P)
            optimal_k = self.crop_means.loc[crop].iloc[2]  # Third column (K)
        except (KeyError, IndexError) as e:
            print(f"Error accessing nutrient data for {crop}: {e}")
            return 0
        
        # Calculate deficiencies (max with 0 to avoid negative values)
        n_deficiency = max(0, optimal_n - soil_n)
        p_deficiency = max(0, optimal_p - soil_p)
        k_deficiency = max(0, optimal_k - soil_k)
        
        # Calculate costs
        n_cost = n_deficiency * self.fertilizer_costs['N']
        p_cost = p_deficiency * self.fertilizer_costs['P']
        k_cost = k_deficiency * self.fertilizer_costs['K']
        
        return n_cost + p_cost + k_cost
    
    def calculate_economics(self, suitable_crops, soil_n=None, soil_p=None, soil_k=None):
        """
        Calculate ROI and profits for suitable crops
        
        Parameters:
        - suitable_crops: List of crops deemed suitable by the existing model
        - soil_n, soil_p, soil_k: Optional soil nutrient levels for fertilizer calculations
        
        Returns:
        - DataFrame with ROI, profits, and other economic metrics sorted by ROI
        """
        results = []
        
        for crop in suitable_crops:
            if crop not in self.average_yields:
                continue
                
            # Calculate revenue
            yield_per_hectare = self.average_yields[crop]
            price_per_unit = self.market_prices[crop]
            
            # Special case for coconut which is measured in nuts, not kg
            if crop == 'coconut':
                revenue = yield_per_hectare * price_per_unit
            else:
                revenue = yield_per_hectare * 1000 * price_per_unit  # Convert tons to kg
            
            # Calculate costs
            seed_cost = self.seed_costs[crop]
            maintenance_cost = self.maintenance_costs[crop]
            
            # Calculate fertilizer costs if soil data is provided
            fertilizer_cost = 0
            if all(x is not None for x in [soil_n, soil_p, soil_k]):
                fertilizer_cost = self.calculate_fertilizer_cost(crop, soil_n, soil_p, soil_k)
            
            total_cost = seed_cost + maintenance_cost + fertilizer_cost
            
            # Calculate profit
            profit = revenue - total_cost
            
            # Calculate ROI
            roi = (profit / total_cost) * 100
            
            # Calculate monthly ROI for fair comparison between crops with different growing periods
            monthly_roi = roi / self.growing_period[crop]
            
            # Calculate profit per month
            monthly_profit = profit / self.growing_period[crop]
            
            results.append({
                'crop': crop,
                'yield_per_hectare': yield_per_hectare,
                'revenue': revenue,
                'seed_cost': seed_cost,
                'maintenance_cost': maintenance_cost,
                'fertilizer_cost': fertilizer_cost,
                'total_cost': total_cost,
                'profit': profit,
                'roi': roi,
                'growing_period_months': self.growing_period[crop],
                'monthly_roi': monthly_roi,
                'monthly_profit': monthly_profit
            })
        
        # Convert to DataFrame and sort by ROI
        results_df = pd.DataFrame(results)
        
        if results_df.empty:
            return results_df
            
        # Add ranking columns
        results_df['roi_rank'] = results_df['roi'].rank(ascending=False)
        results_df['profit_rank'] = results_df['profit'].rank(ascending=False)
        results_df['monthly_roi_rank'] = results_df['monthly_roi'].rank(ascending=False)
        results_df['monthly_profit_rank'] = results_df['monthly_profit'].rank(ascending=False)
        
        # Calculate an overall score (lower is better)
        results_df['overall_score'] = (
            results_df['roi_rank'] * 0.3 + 
            results_df['profit_rank'] * 0.3 + 
            results_df['monthly_roi_rank'] * 0.2 + 
            results_df['monthly_profit_rank'] * 0.2
        )
        
        # Sort by overall score (ascending since lower score is better)
        results_df = results_df.sort_values('overall_score')
        
        # Format monetary values for display
        for col in ['revenue', 'seed_cost', 'maintenance_cost', 'fertilizer_cost', 'total_cost', 'profit', 'monthly_profit']:
            results_df[col] = results_df[col].round(2)
            
        # Format percentage values
        for col in ['roi', 'monthly_roi']:
            results_df[col] = results_df[col].round(2)
            
        return results_df
    
    def get_recommendations(self, suitable_crops, soil_params=None, top_n=22):
        """
        Get crop recommendations sorted by economic factors
        
        Parameters:
        - suitable_crops: List of crops from existing suitability model
        - soil_params: Optional dictionary with 'N', 'P', 'K' values
        - top_n: Number of top recommendations to return
        
        Returns:
        - DataFrame with top N recommendations
        """
        soil_n = soil_params.get('N') if soil_params else None
        soil_p = soil_params.get('P') if soil_params else None
        soil_k = soil_params.get('K') if soil_params else None
        
        economics_df = self.calculate_economics(suitable_crops, soil_n, soil_p, soil_k)
        
        if economics_df.empty:
            return pd.DataFrame()
            
        # Get top N recommendations
        top_recommendations = economics_df.head(top_n)
        
        return top_recommendations
    
    def get_detailed_report(self, crop_name, soil_params=None):
        """
        Generate a detailed economic report for a specific crop
        
        Parameters:
        - crop_name: Name of the crop to analyze
        - soil_params: Optional dictionary with 'N', 'P', 'K' values
        
        Returns:
        - Dictionary with detailed economic analysis
        """
        soil_n = soil_params.get('N') if soil_params else None
        soil_p = soil_params.get('P') if soil_params else None
        soil_k = soil_params.get('K') if soil_params else None
        
        economics_df = self.calculate_economics([crop_name], soil_n, soil_p, soil_k)
        
        if economics_df.empty:
            return {"error": f"Crop {crop_name} not found in database"}
            
        crop_data = economics_df.iloc[0].to_dict()
        
        # Add additional insights
        if self.crop_means is not None:
            try:
                crop_data['optimal_conditions'] = {
                    'N': self.crop_means.loc[crop_name].iloc[0],
                    'P': self.crop_means.loc[crop_name].iloc[1],
                    'K': self.crop_means.loc[crop_name].iloc[2],
                    'rainfall_mm': self.crop_means.loc[crop_name].iloc[3],
                    'pH': self.crop_means.loc[crop_name].iloc[4],
                    'humidity_%': self.crop_means.loc[crop_name].iloc[5],
                    'temperature_C': self.crop_means.loc[crop_name].iloc[6]
                }
            except (KeyError, IndexError) as e:
                print(f"Error accessing data for {crop_name}: {e}")
                crop_data['optimal_conditions'] = {}
            
        return crop_data


# Function to predict suitable crops using the RF model
def predict_suitable_crops(rf_model, soil_data):
    """
    Use the Random Forest model to predict suitable crops based on soil and climate data
    
    Parameters:
    - rf_model: The trained Random Forest classifier
    - soil_data: Dictionary with soil and climate parameters
    
    Returns:
    - List of suitable crops
    """
    features = ['N', 'P', 'K', 'ph', 'temperature', 'humidity', 'rainfall']
    
    # Create input data from the soil_data dictionary
    input_values = [
        soil_data.get('N', 0),
        soil_data.get('P', 0),
        soil_data.get('K', 0),
        soil_data.get('pH', 0),  # Note: case sensitive, matches your soil_data dictionary
        soil_data.get('temperature', 0),
        soil_data.get('humidity', 0),
        soil_data.get('rainfall', 0)
    ]
    
    input_data = pd.DataFrame([input_values], columns=features)
    
    # Define all possible crops in the same order as used in model training
    crops = ['apple', 'banana', 'blackgram', 'chickpea', 'coconut', 'coffee',
             'cotton', 'grapes', 'jute', 'kidneybeans', 'lentil', 'maize',
             'mango', 'mothbeans', 'mungbean', 'muskmelon', 'orange', 'papaya',
             'pigeonpeas', 'pomegranate', 'rice', 'watermelon']
    
    # Option 1: If your model only predicts the most likely crop class
    # prediction = rf_model.predict(input_data)
    # return [crops[prediction[0]]]
    
    # Option 2: If your model can provide probabilities, return multiple suitable crops
    try:
        # Try to get probabilities for all classes
        probabilities = rf_model.predict_proba(input_data)[0]
        
        # Find crops with probability above a threshold (e.g., 0.05 or 5%)
        threshold = 0.05
        suitable_crops = [crops[i] for i, prob in enumerate(probabilities) if prob > threshold]
        
        # If no crops meet the threshold, return the top 5 most likely ones
        if not suitable_crops:
            top_indices = probabilities.argsort()[-5:][::-1]  # Get indices of top 5 probabilities
            suitable_crops = [crops[i] for i in top_indices]
            
        return suitable_crops
        
    except Exception as e:
        print(f"Error getting probabilities: {e}")
        # Fallback: For testing or if predict_proba is not available, return all crops
        return crops


# Function to integrate the RF model with economic analysis
def integrate_with_existing_model(rf_model, soil_data, recommender, top_n=22):
    """
    Integrate the economic recommender with the Random Forest model
    
    Parameters:
    - rf_model: The RandomForest classifier model
    - soil_data: Dictionary with soil and climate data (N, P, K, pH, temp, humidity, rainfall)
    - recommender: An instance of CropRecommender
    - top_n: Number of top recommendations to return
    
    Returns:
    - DataFrame with top N economically viable crops that are suitable for the conditions
    """
    # Get suitable crops from the RF model
    suitable_crops = predict_suitable_crops(rf_model, soil_data)
    
    print(f"Crops suitable for the given conditions: {suitable_crops}")
    
    # Get soil parameters for fertilizer calculations
    soil_params = {
        'N': soil_data.get('N'),
        'P': soil_data.get('P'),
        'K': soil_data.get('K')
    }
    
    # Get economic recommendations from the suitable crops
    recommendations = recommender.get_recommendations(suitable_crops, soil_params, top_n)
    
    return recommendations


# Function to create the recommender with your data
def create_recommender_from_data():
    """
    Create and initialize the crop recommender with economic data
    """
    # Define your economic data here or load it from files
    # Example dictionaries:
    
    # Average yield in tons per hectare
    average_yields = {
        'rice': 4.0, 'maize': 3.5, 'chickpea': 1.2, 'kidneybeans': 1.1, 'pigeonpeas': 1.0,
        'mothbeans': 0.9, 'mungbean': 1.0, 'blackgram': 0.9, 'lentil': 1.0, 'pomegranate': 15.0,
        'banana': 35.0, 'mango': 10.0, 'grapes': 18.0, 'watermelon': 30.0, 'muskmelon': 25.0,
        'apple': 20.0, 'orange': 15.0, 'papaya': 40.0, 'coconut': 12000, 'cotton': 2.0,
        'jute': 2.5, 'coffee': 1.0
    }
    
    # Market prices in ₹ per kg (or per nut for coconut)
    market_prices = {
        'rice': 20, 'maize': 15, 'chickpea': 60, 'kidneybeans': 80, 'pigeonpeas': 70,
        'mothbeans': 60, 'mungbean': 65, 'blackgram': 70, 'lentil': 75, 'pomegranate': 80,
        'banana': 25, 'mango': 50, 'grapes': 60, 'watermelon': 15, 'muskmelon': 20,
        'apple': 70, 'orange': 40, 'papaya': 30, 'coconut': 15, 'cotton': 60,
        'jute': 40, 'coffee': 300
    }
    
    # Seed costs in ₹ per hectare
    seed_costs = {
        'rice': 5000, 'maize': 4500, 'chickpea': 6000, 'kidneybeans': 7000, 'pigeonpeas': 5500,
        'mothbeans': 5000, 'mungbean': 5200, 'blackgram': 5000, 'lentil': 6500, 'pomegranate': 50000,
        'banana': 35000, 'mango': 40000, 'grapes': 80000, 'watermelon': 8000, 'muskmelon': 7000,
        'apple': 100000, 'orange': 50000, 'papaya': 15000, 'coconut': 25000, 'cotton': 10000,
        'jute': 6000, 'coffee': 70000
    }
    
    # Maintenance costs in ₹ per hectare
    maintenance_costs = {
        'rice': 15000, 'maize': 12000, 'chickpea': 10000, 'kidneybeans': 12000, 'pigeonpeas': 10000,
        'mothbeans': 8000, 'mungbean': 9000, 'blackgram': 9000, 'lentil': 9500, 'pomegranate': 80000,
        'banana': 60000, 'mango': 50000, 'grapes': 120000, 'watermelon': 25000, 'muskmelon': 20000,
        'apple': 150000, 'orange': 60000, 'papaya': 40000, 'coconut': 30000, 'cotton': 25000,
        'jute': 15000, 'coffee': 100000
    }
    
    # Fertilizer costs in ₹ per kg/unit
    fertilizer_costs = {
        'N': 20,  # Cost per kg of Nitrogen
        'P': 30,  # Cost per kg of Phosphorous
        'K': 25   # Cost per kg of Potassium
    }
    
    # Growing period in months
    growing_period = {
        'rice': 4, 'maize': 4, 'chickpea': 4, 'kidneybeans': 3, 'pigeonpeas': 5,
        'mothbeans': 3, 'mungbean': 3, 'blackgram': 4, 'lentil': 5, 'pomegranate': 36,
        'banana': 12, 'mango': 60, 'grapes': 24, 'watermelon': 3, 'muskmelon': 3,
        'apple': 60, 'orange': 36, 'papaya': 12, 'coconut': 72, 'cotton': 6,
        'jute': 4, 'coffee': 36
    }
    
    # Create the recommender with the data
    recommender = CropRecommender(
        average_yields=average_yields,
        market_prices=market_prices,
        seed_costs=seed_costs,
        maintenance_costs=maintenance_costs,
        fertilizer_costs=fertilizer_costs,
        growing_period=growing_period
    )
    
    # Create or load crop_means data
    # This should be a DataFrame with crop names as the index and columns for optimal values
    # Example (you should replace this with your actual data):
    crop_data = {
        'rice': [78.89, 47.58, 38.97, 236.18, 6.42, 82.27, 23.68],
        'maize': [77.76, 48.44, 19.79, 84.77, 6.24, 65.09, 22.38],
        'apple': [20.80, 134.22, 199.89, 112.65, 5.93, 92.33, 22.63],
        'banana': [100.23, 82.01, 50.05, 104.62, 5.98, 80.35, 27.37],
        'blackgram': [40.02, 67.47, 19.24, 67.88, 7.13, 65.11, 29.97],
        'chickpea': [40.09, 67.79, 79.92, 80.05, 7.33, 16.86, 18.87],
        'coconut': [21.98, 16.93, 30.59, 175.68, 5.97, 94.84, 27.40],
        'coffee': [101.20, 28.74, 29.94, 158.07, 6.79, 58.87, 25.54],
        'cotton': [117.77, 46.24, 19.56, 80.39, 6.91, 79.84, 23.99],
        'grapes': [23.18, 132.53, 200.11, 69.61, 6.02, 81.87, 23.85],
        'jute': [78.40, 46.86, 39.99, 174.79, 6.73, 79.63, 24.95],
        'kidneybeans': [20.75, 67.54, 20.05, 105.92, 5.74, 21.60, 20.11],
        'lentil': [18.77, 68.36, 19.41, 45.68, 6.92, 64.80, 24.50],
        'mango': [20.07, 27.18, 29.92, 94.70, 5.76, 50.16, 31.20],
        'mothbeans': [21.44, 48.01, 25.19, 51.19, 6.83, 53.16, 18.19],
        'mungbean': [20.09, 47.28, 19.87, 48.40, 6.72, 85.50, 28.52],
        'muskmelon': [100.32, 17.72, 50.08, 24.68, 6.53, 92.34, 28.66],
        'orange': [19.58, 16.55, 10.01, 110.47, 7.01, 92.17, 22.76],
        'papaya': [49.88, 59.05, 50.04, 142.62, 6.74, 92.40, 33.72],
        'pigeonpeas': [20.73, 67.73, 20.29, 149.45, 5.79, 48.06, 27.74],
        'pomegranate': [18.87, 17.85, 30.91, 107.52, 6.42, 90.12, 21.83],
        'watermelon': [99.42, 17.00, 50.22, 50.78, 6.49, 85.16, 25.59]
    }

    
    # Convert to DataFrame with columns for N, P, K, rainfall, pH, humidity, temperature
    crop_means = pd.DataFrame.from_dict(
        crop_data, 
        orient='index',
        columns=['N', 'P', 'K', 'rainfall', 'pH', 'humidity', 'temperature']
    )
    
    # Set the crop means data
    recommender.set_crop_means(crop_means)
    
    return recommender


# Function to format results for display
def format_results_for_display(recommendations):
    """
    Format the recommendations for display
    
    Parameters:
    - recommendations: DataFrame with crop recommendations
    
    Returns:
    - DataFrame with formatted columns for display
    """
    if recommendations.empty:
        return pd.DataFrame()
        
    display_df = recommendations[['crop', 'yield_per_hectare', 'revenue', 
                                 'total_cost', 'profit', 'roi', 
                                 'growing_period_months', 'monthly_roi']].copy()
    
    # Format monetary values as Indian Rupees
    for col in ['revenue', 'total_cost', 'profit']:
        display_df[col] = display_df[col].apply(lambda x: f"₹{x:,.2f}")
    
    # Format percentages
    display_df['roi'] = display_df['roi'].apply(lambda x: f"{x:.2f}%")
    display_df['monthly_roi'] = display_df['monthly_roi'].apply(lambda x: f"{x:.2f}%")
    
    # Rename columns for clarity
    display_df.columns = ['Crop', 'Yield (tons/ha)', 'Revenue', 'Total Cost', 
                          'Profit', 'ROI', 'Growing Period (months)', 'Monthly ROI']
    
    return display_df


# Main function for the complete workflow
def main():
    """
    Main function to run the complete crop recommendation system
    """
    # Load your RandomForest model
    # Uncomment and modify path as needed
    # with open('your_model_path.pkl', 'rb') as f:
    #     rf_model = pickle.load(f)
    
    # For testing without the actual model, use a dummy model
    with open('random_forest_classifier.pkl', 'rb') as f:
        rf_model = pickle.load(f)
    # Create the recommender with economic data
    recommender = create_recommender_from_data()
    
    # Get soil and climate data (from user input or sensors)
    soil_data = {
        'N': 80,
        'P': 40, 
        'K': 20,
        'pH': 6.5,
        'temperature': 25,
        'humidity': 80,
        'rainfall': 200
    }
    
    # Get recommendations by integrating RF model and economics
    recommendations = integrate_with_existing_model(
        rf_model=rf_model,
        soil_data=soil_data,
        recommender=recommender,
        top_n=10 # Top 5 most economically viable from all suitable crops
    )
    
    # Format and display results
    formatted_results = format_results_for_display(recommendations)
    print("Top Crop Recommendations by Economic Factors:")
    print(formatted_results)
    
    # Get detailed report for the top recommended crop
    if not recommendations.empty:
        top_crop = recommendations.iloc[0]['crop']
        detailed_report = recommender.get_detailed_report(top_crop, soil_params=soil_data)
        
        print(f"\nDetailed Analysis for {top_crop.capitalize()}:")
        for key, value in detailed_report.items():
            if key != 'optimal_conditions':
                print(f"{key}: {value}")
        
        print("\nOptimal Growing Conditions:")
        for key, value in detailed_report.get('optimal_conditions', {}).items():
            print(f"{key}: {value}")


if __name__ == "__main__":
    main()
```
